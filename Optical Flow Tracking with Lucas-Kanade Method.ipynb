This notebook implements optical flow tracking using the Lucas-Kanade method in OpenCV. It reads a video file (coastguard_qcif.y4m), extracts frames, and detects key feature points using cv2.goodFeaturesToTrack(). The motion of these points is then tracked between consecutive frames using cv2.calcOpticalFlowPyrLK(). The results are visualized by overlaying motion vectors on the frames. 

```python
video_path = "coastguard_qcif.y4m"

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('frame', frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

video_path = "coastguard_qcif.y4m"

cap = cv2.VideoCapture(video_path)

ret, prev_frame = cap.read()

if not ret:
    print("Error")
    cap.release()
    cv2.destroyAllWindows()
else:
    ret, next_frame = cap.read()

    if not ret:
        print("Errot")
        cap.release()
        cv2.destroyAllWindows()
    else:
        prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
        next_gray = cv2.cvtColor(next_frame, cv2.COLOR_BGR2GRAY)

        prev_pts = cv2.goodFeaturesToTrack(prev_gray, maxCorners=100, qualityLevel=0.3, minDistance=7)

        lk_params = dict(winSize=(15, 15), maxLevel=5, criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

        next_pts, status, error = cv2.calcOpticalFlowPyrLK(prev_gray, next_gray, prev_pts, None, **lk_params)

        for i, (new, old) in enumerate(zip(next_pts, prev_pts)):
            a, b = new.ravel().astype(int)
            c, d = old.ravel().astype(int)
            next_frame = cv2.line(next_frame, (a, b), (c, d), (0, 255, 0), 2)
            next_frame = cv2.circle(next_frame, (a, b), 5, (0, 255, 0), -1)

        cv2.imshow('Optical Flow', next_frame)
        cv2.waitKey(30000)

        cap.release()
        cv2.destroyAllWindows()
```
